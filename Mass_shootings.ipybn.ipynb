```python
import pandas as pd
import os
import csv
import json
import pymongo
from collections import defaultdict
```

```python
data = pd.read_csv("datasets/Mother Jones' Investigation_ US Mass Shootings, 1982-2018 - US mass shootings.csv")
data
```

```python
lat_lon_data = pd.read_csv("datasets/correct_lat_lng_mass_shootings_final_20171116_835.csv")
lat_lon_data
```

```python
combined_dataset = data.merge(lat_lon_data,left_on='Location', right_on='Location', how='outer')

combined_dataset['Year'] = combined_dataset["Year"].fillna(0).astype(int)
combined_dataset['Total victims'] = combined_dataset["Total victims"].fillna(0).astype(int)
combined_dataset['Fatalities'] = combined_dataset["Fatalities"].fillna(0).astype(int)
combined_dataset['Injured'] = combined_dataset["Injured"].fillna(0).astype(int)
combined_dataset
```

```python
clean_df = combined_dataset.drop(["latitude", "longitude"], axis = 1)
clean_df
```

```python
clean_df.to_csv("Mass_shootings_data.csv")
```

```python
conn = 'mongodb://localhost:27017'
client = pymongo.MongoClient(conn)

db = client.us_mass_shootings
collection = db.incidents
```

```python
year_list = []
events = list(collection.find())
for event in events:
    if event["Year"] != 0:
        year_list.append(event["Year"])
print(year_list)
```

```python
d = defaultdict(int)

for year in year_list:
    d[year] += 1
d
```

```python
number_of_victims = []
for event in events:
    victims_per_year = {}
    if event["Year"] != 0:
        victims_per_year["Year"] = event["Year"]
        victims_per_year["Fatalities"] = event["Fatalities"]
        victims_per_year["Injured"] = event["Injured"]        
        number_of_victims.append(victims_per_year)
number_of_victims

    
    

    

    
```
